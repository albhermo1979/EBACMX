```python
#Hacer un programa que permita al usuario solicitar la tabla de multiplicar del numero que quiera
def tabla_del(numero):
    resultados = []
    for i in range(1,11):
        resultados.append(numero * i) # Append mete los valores a la lista
    return resultados

for valor in range(1,11):
    print(tabla_del(valor))
    
```

```python
# misma funcion pero con la validacion de mayusculas o minusculas
def palindromo(cadena):
    izquierda = 0
    derecha = len(cadena) - 1

    while derecha >= izquierda:
        if not(cadena[izquierda].lower() == cadena[derecha].lower()):
            return false
        izquierda  += 1
        derecha -= 1
    return True

print(palindromo("RayaR"))
```

```python
# Suma sin recuersividad
def suma(lista):
    suma = 0
    for i in range(0, len(lista)):
        suma = suma + lista[i]
    return suma

print(suma([6,3,4,2,10]))
```

```python
# Funcion recursiva
def suma(lista):
    if len(lista) == 1:
        return(lista[0])
    else:
        return lista[0] + suma(lista[1:])

print(suma([6,3,4,2,10]))
```

```python
#Recursividad del factorial de un numero
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(10))
```

```python
def invierte_cadena(cadena):
    reversa = []
    longitud = len(cadena) - 1
    while longitud >= 0:
        reversa.append(cadena[longitud])
        longitud -= 1
    return reversa
    
print(invierte_cadena("RayaR2"))
```

```python
def invierte_cadena(cadena):
    reversa = ""
    for caracter in cadena:
        reversa = caracter + reversa
    return reversa
    
invierte_cadena("123abc")
```

```python
def prueba_primo(numero):
    if numero == 1:
        return True
    else:
        for i in range(1,numero):
            primo = numero / i
            if primo == 1:
            return True
        else:
            return False

prueba_primo(4)
```

```python
def prueba_primo(numero):
    if numero <= 1:
        return False
    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            return False
    return True

prueba_primo(5)
```
