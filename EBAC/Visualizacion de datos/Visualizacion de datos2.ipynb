```python
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv('co2-mm-mlo_csv.csv')
df
```

```python
#Aca vamos a sacar la informacion del DataFrame
fechas = df[df["Date"] > "2015-01-01"]
fig, ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
plt.show
```

```python
#Para que se quiten las labels de los datos pra una mejor visualizacion
fechas = df[df["Date"] > "2015-01-01"]
fig, ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
ax.axes.xaxis.set_ticks([]) #Con esta se eliminan las labels de los ejes
plt.xlabel("Datos mensuales: 2015-10-01 - 2018-09-01")
plt.ylabel("Emisiones de CO2 promedio")
plt.show()
```

```python
#Crear una funcion que grafica series de tiempo
def grafica_series_tiempo(ejes, x, y, color, etiqueta_x, etiqueta_y):
    ejes.plot(x,y, color = color)
    ejes.set_xlabel(etiqueta_x)
    ejes.set_ylabel(etiqueta_y, color = color)
    ejes.tick_params('y', color = color)
```

```python
fig, ax = plt.subplots()
grafica_series_tiempo(ax, fechas["Date"], fechas["Interpolated"], 'blue', 'Datos menusales: 2015-01-01 - 2018-09-01', 'CO2')
ax.axes.xaxis.set_ticks([]) #Con esta se eliminan las labels de los ejes
plt.show()
```

```python
#Grafica con doble ejes
fig, ax = plt.subplots()
grafica_series_tiempo(ax, fechas["Date"], fechas["Interpolated"], 'blue', 'Datos menusales: 2015-01-01 - 2018-09-01', 'CO2')

ax2 = ax.twinx()
grafica_series_tiempo(ax2, fechas["Date"], fechas["Trend"], 'red', '', 'CO2 Trend')
ax.axes.xaxis.set_ticks([]) #Con esta se eliminan las labels de los ejes
plt.show()
```
