```python
#Con esto le estoy diciendo a Python que el indice va a ser la columna cero o sea los paises
df = pd.read_csv('olympic-medals.csv', index_col = 0)
df
```

```python
#Tomamos a los 10 equipos que ganaron mas medallas, como el DataFrame esta ordenado solo tomamos los primeros 10
medallas = df.head(10)
medallas
```

```python
#Visualizacion de medallas de Oro y Plata
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
#Grafico de barras apiladas
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom=medallas["Gold"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
#Grafico de barras apiladas
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom=medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom=medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
#Grafico de barras apiladas
fig, ax = plt.subplots()
ax.bar(medallas.index, medallas["Gold"])
ax.bar(medallas.index, medallas["Silver"], bottom=medallas["Gold"])
ax.bar(medallas.index, medallas["Bronze"], bottom=medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index, rotation=90)
ax.set_ylabel("Numero de medallas")
ax.legend(["Gold", "Silver", "Bronze"])
plt.show()
```

```python
#Creacion de un histograma para mostrar de todos los paises cuantas medallas de oro obtienen
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.hist(df["Gold"])
plt.show()
```

```python
#Creacion de un histograma para mostrar de todos los paises cuantas medallas de oro y plata obtienen
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro")
ax.hist(df["Silver"], label = "Plata")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show()
```

```python
#Creacion de un histograma con especificacion de intervalos
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro", bins = [0, 5, 10, 15, 20, 25, 30, 35, 40])
ax.hist(df["Silver"], label = "Plata", bins = [0, 5, 10, 15, 20, 25, 30, 35, 40])
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show()
```

```python
#Creacion de un histograma con especificacion de intervalos y trans[arencias
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.hist(df["Gold"], label = "Oro", bins = [0, 5, 10, 15, 20, 25, 30, 35, 40], histtype = "step")
ax.hist(df["Silver"], label = "Plata", bins = [0, 5, 10, 15, 20, 25, 30, 35, 40], histtype = "step")
ax.set_xlabel("# de medallas")
ax.set_ylabel("# de observaciones")
ax.legend()
plt.show()
```

```python
#Creacion de diagramas de caja
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.boxplot([df["Gold"], df["Silver"]])
ax.set_xticklabels(["Oro", "Plata"])
ax.set_ylabel("# de medallas")
plt.show()
```

```python
#Creacion de grafico de dispersion
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"])
ax.set_xlabel("# medallas de Oro")
ax.set_ylabel("# de medallas Plata")
plt.show()
```

```python
#Creacion de grafico de dispersion multiples
import matplotlib.pyplot as plt # Se define la librearia de graficacion
fig, ax = plt.subplots()
ax.scatter(df["Gold"], df["Silver"], color = "red", label = "Plata")
ax.scatter(df["Gold"], df["Bronze"], color = "blue", label = "Bronce")
ax.set_xlabel("# medallas de Oro")
ax.set_ylabel("# de medallas Plata")
ax.legend()
plt.show()
```
