```python
import matplotlib.pyplot as plt
#from Matplotlib import pyplot as plt # Otra forma de importar pyplot
import numpy as np

# Datos de ejemplo
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
temperaturas = [22, 21, 23, 24, 25]

# Crear gráfico de líneas
plt.plot(meses, temperaturas, marker='o', markerfacecolor='white', markeredgecolor='black', linestyle='-', color='b')
plt.title('Temperaturas Mensuales')
plt.xlabel('Meses')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')
df
```

```python
fig, ax = plt.subplots()
hist = ax.boxplot(df["Age"], bins=10)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
hist
```

```python
# A partir de la columna Age, que contiene la edad de cada pasajero, obtenga el diagrama de caja para la variable.
# Realice los ajustes que considere necesarios al DataFrame.
# Almacenar el gráfico en la variable box.
fig, ax = plt.subplots()
hist = ax.boxplot(df["Age"])
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
hist
```

```python

grouped_df = df.groupby('Pclass')[['Age']].count()
grouped_df
```

```python
fig, ax = plt.subplots()
bar = ax.bar(grouped_df.index, grouped_df["Age"])
plt.show()
```
