```python
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
#Analisis con histogramas con pandas de manera rapida
#En pandas el histograma se muestra con frecuencias absolutas, sin etiqueta automatica para eje x e intervalos amplios
df["math score"].plot.hist()
#Nos muestra que la distibucion de los datos esta entre el 60 y 70
```

```python
#Atraves de Seaborn se hace de la siguiente manera el histograma
#En seaborn las fecuencias son relativas, con intervaloes estrechos
import seaborn as sns
sns.distplot(df["math score"])
```

```python
#Histograma alternatico en Seaborn
sns.displot(df["math score"])
```

```python
#Histograma alternatico en Seaborn con mas opciones
sns.displot(df["math score"], kde = True, bins = 10) # Con la curva de Densidad
```

```python
sns.distplot(df["math score"], hist = False, kde_kws = {'shade': True}) # Con la curva de Densidad
```

```python
#Esto en con la nueva version
sns.kdeplot(df["math score"])
```

```python
sns.distplot(df["math score"], hist = False, kde_kws = {'shade': True}) # Con la curva de Densidad
```

```python
#Creacion de digramas de dispersion con SeaBorn con recta de regresion
sns.regplot(x="reading score", y="writing score", data=df)
```

```python
sns.lmplot(x="reading score", y="writing score", data=df)
```

```python
#Grafica de dispersion con regresion lineal por categorias de genero
sns.lmplot(x="reading score", y="writing score", data=df, hue = "gender")
```

```python
#Grafica de dispersion con regresion lineal por categorias de genero separado en 2 partes
sns.lmplot(x="reading score", y="writing score", data=df, col = "gender")
```

```python

sns.lmplot(x="reading score", y="writing score", data=df, col = "race/ethnicity")
```
