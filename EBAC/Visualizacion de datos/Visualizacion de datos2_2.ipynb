```python
import pandas as pd
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
#Estilos en SeaBorn
import seaborn as sns
for style in ["dark", "whitegrid", "darkgrid", "ticks"]:
    sns.set_style(style)
    sns.displot(df["reading score"])
```

```python
#Remocion de ejes
sns.set_style("ticks")
sns.displot(df["reading score"], kde = True)
sns.despine(left=True, bottom =True)
```

```python
#Combinacion de parametros con Matplotlib
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax = ax)
ax.set(xlabel = "Score de escritura", ylabel = "Frecuencia relativa")
```

```python
#Truncamiento de esacala de visualizacion
fig, ax = plt.subplots()
sns.distplot(df["writing score"], ax = ax)
ax.set(xlabel = "Score de escritura", ylabel = "Frecuencia relativa", xlim = (0, 90), title = "Histograma de frecuencia para puntajes de escritura")
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2, sharey = True)
sns.distplot(df["writing score"], ax = ax0) #Un histograma para TODAS las observaciones
sns.distplot(df.query('gender == "male"')["writing score"], ax = ax1) #Histrograma para los hombres
ax0.set(xlabel = "Puntaje general de escritura")
ax1.set(xlabel = "Puntaje general de escritura para hombres")
ax1.axvline(x = 75, label = "Objetivo", linestyle = "--")
ax1.legend()
```

```python
# Asignacion de colores al histrograma
paleta = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]
for p in paleta:
    sns.set_palette(p)
    sns.displot(df["reading score"], kde = True)
```
