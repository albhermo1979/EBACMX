```python
import pandas as pd
df = pd.read_csv('The_Population.csv')
df
```

```python
df['Value'] = pd.to_numeric(df['Value'].str.replace(',', ''))
df
```

```python
df_2020 = df[df['Year'] == 2020]
```

```python
df_pivotado = df_2020.pivot_table(index=['Region', 'Country/Area', 'Year'], columns='Series', values='Value')
df_pivotado
```

```python
correlacion = df_pivotado.corr()
correlacion
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', square=True)
plt.title('Heatmap de Correlación')
plt.show()
```
