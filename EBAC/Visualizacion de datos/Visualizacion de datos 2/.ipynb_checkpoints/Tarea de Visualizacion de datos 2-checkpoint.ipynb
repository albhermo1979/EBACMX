```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('The_Population.csv')
df
```

```python
#Convierto la columna Value a numerico para que no me de ningun problema y le quito las comas
df['Value'] = pd.to_numeric(df['Value'].str.replace(',', ''))
```

```python
#Hago un frltrado solo de los datos del 2020
df_2020 = df[df['Year'] == 2020]

#Hago una separacion de los valores de mortalidad infantil y esperanza de vida en un Data_Frame llamado df_pivot
df_pivot = df_2020[df_2020['Series'].isin(['Infant mortality for both sexes (per 1,000 live births)', 
                                           'Life expectancy at birth for both sexes (years)'])]

#Aca hago la tabla pivote con el indice el Pais, la columna series y los valores, uti
df_pivot = df_pivot.pivot_table(index='Country/Area', columns='Series', values='Value')

plt.figure(figsize=(10, 6))
sns.regplot(x='Infant mortality for both sexes (per 1,000 live births)', 
                y='Life expectancy at birth for both sexes (years)', 
                data=df_pivot, x_estimator = np.mean)

# Configuraciones adicionales
plt.title('RelaciÃ³n entre mortalidad infantil y esperanza de vida en 2020')
plt.xlabel('Tasa de mortalidad infantil')
plt.ylabel('Esperanza de vida')

# Mostrar el grÃ¡fico
plt.tight_layout()
plt.show()
```
