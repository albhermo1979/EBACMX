```python
#Uso del estilo ggplot
import pandas as pd
import matplotlib.pyplot as plt 
plt.style.use("ggplot") #Con esto se le da un color de fondo
df = pd.read_csv('co2-mm-mlo_csv.csv')
fechas = df[df["Date"] > "2015-01-01"]
fig, ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
ax.axes.xaxis.set_ticks([]) #Con esta se eliminan las labels de los ejes
plt.xlabel("Datos mensuales: 2015-10-01 - 2018-09-01")
plt.ylabel("Emisiones de CO2 promedio")
plt.show()
```

```python
#Con eso se regresa al estilo original
plt.style.use("default") #Con esto se le da un color de fondo
df = pd.read_csv('co2-mm-mlo_csv.csv')
fechas = df[df["Date"] > "2015-01-01"]
fig, ax = plt.subplots()
ax.plot(fechas["Date"], fechas["Interpolated"])
ax.axes.xaxis.set_ticks([]) #Con esta se eliminan las labels de los ejes
plt.xlabel("Datos mensuales: 2015-10-01 - 2018-09-01")
plt.ylabel("Emisiones de CO2 promedio")
plt.show()
```
