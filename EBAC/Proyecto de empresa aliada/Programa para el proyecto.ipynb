```python
import pandas as pd
df_fact = pd.read_csv('FACT_SALES.csv')
df_prod = pd.read_csv('DIM_PRODUCTOS.csv')
```

```python
df_prod.head()
```

```python
df_fact.head()
```

```python
#Asi se renombra la columna
df_prod = df_prod.rename(columns={'ITEM': 'ITEM_CODE'})
```

```python
df_prod.head()
```

```python
df_unido = pd.merge(df_fact, df_prod, on="ITEM_CODE")
df_unido.head()
```

```python
# Aplicar función para remover ITEM_CODE de DESCRIPCION
def remover_item_code(row):
    return row['ITEM_DESCRIPTION'].replace(row['ITEM_CODE'], '').strip()

df_unido['ITEM_DESCRIPTION'] = df_unido.apply(remover_item_code, axis=1)
```
