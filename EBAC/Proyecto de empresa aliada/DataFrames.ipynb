```python
import pandas as pd
dfp = pd.DataFrame([['Bella','Labrador','Cafe', 56, 24, '2013-07-01'], ['Charlie','Poodle','Negro', 43, 24, '2016-09-16'], ['Lucy','Chow Chow','Café', 46, 24, '2014-08-35'], ['Cooper','Schnauzer','Gris', 49, 17, '2011-12-11'], ['Bernie','San Bernardo','Blanco', 77, 74, '2018-02-27']], index = [0, 1, 2, 3, 4], columns = ["Nombre", "Raza", "Color", "Altura_CM", "Peso_KG", "Fecha_Nac"])
dfp
```

```python
#Para extraer los valores de aquellos perros que nacieron antes del 2015-01-01
dfp2 = dfp[dfp['Fecha_Nac'] < '2015-01-01']
dfp2
```

```python
#Si un campo numerico se tomara como String, se puede hacer el cambio con el siguiente comando:
# dfp['Altura_CM'] = dfp['Altura_CM'].astype(int)
dfp3 = dfp[dfp['Altura_CM'] > 50]
dfp3
```

```python
dfp["IMM"] = dfp['Peso_KG'] / (dfp['Altura_CM'] / 100) ** 2
dfp
```

```python
#Ordenar 
dfp 
```

```python
dfp = dfp.sort_values("IMM", ascending = True)
dfp
```
