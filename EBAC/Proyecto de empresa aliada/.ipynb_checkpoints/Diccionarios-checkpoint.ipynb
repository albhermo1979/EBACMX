```python
#Uso de los diccionarios: Parae ver repeticion de palabras en la lista
conteo = dict()
nombres = ['Juan', 'Poncho', 'Poncho', 'Pedro', 'Pancho']
#Para cada nombre en la lista de nombres
#Checar si el nombre ya existe en el diccionario "Conteo"
#Si el nombre no esta en la lista, inicializar su frecuencia en 1
#En caso contrario (ya esta en la lista), incrementar su precuencia en 1
for nombre in nombres:
    if nombre not in conteo:
        conteo[nombre] = 1
    else:
        conteo[nombre] = conteo[nombre] + 1
conteo
```

```python
#Vamos a abrir un archivo de Texto
archivo = open('Trabalenguas.txt')
archivo
```

```python
for texto in archivo:
    palabras = texto.split()
palabras
```

```python
conteo = dict()
for nombre in palabras:
    if nombre not in conteo:
        conteo[nombre] = 1
    else:
        conteo[nombre] = conteo[nombre] + 1
conteo
```
