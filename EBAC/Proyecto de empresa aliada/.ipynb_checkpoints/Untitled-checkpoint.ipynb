```python
#Voy a revisar si existen valores en cero Faltantes por columna = fpc
fpc1 = df_calendar.isnull().sum()
fpc2 = df_category.isnull().sum()
fpc3 = df_product.isnull().sum()
fpc4 = df_segment.isnull().sum()
fpc5 = df_fact_sales.isnull().sum()
print(f"Existen estos faltantes para calendar, {fpc1}")
```
