```python
nombre = input("Nombre del estudiante: ")
datos = {'Jose': 85, "Angela": 92, 'Miguel': 75, 'Ana': 98}
for estudiante in datos:
    if estudiante == nombre:
        print("Calificacion: ", datos[estudiante])
        break
else: 
    print("No se encontro al estudiante solicitado")
        
```

```python
#Ciclo anidado
frutas = ['Manzana', 'Fresa', 'Cereza']
adjetivo = ['Grande', 'Roja', 'Deliciosa']
for fruta in frutas:
    for caracteristica in adjetivo:
        print(fruta, caracteristica)
```

```python
#Para recorrer a la inversa
numeros = [10,20,30,40]
for numero in reversed(numeros):
    print(numero)
    
```

```python
#Ejercicio de las tablas de multiplicar
tablas = int(input("Cuantas tablas quieres calcular:"))
for i in range(1, tablas + 1):
    print(f"La tabla de multiplicar del {i}")
    contador = 1
    while contador < 11:
        print(i * contador, end = ' ') #Con la instruccion end lo que le estoy diciendo es que me imprima un espacio al final
        contador += 1
    print("\n")
    
```
