```python
import pandas as pd
df = pd.read_csv("iris.csv")
df
```

```python
#Se filtra el DataFrame cuando el valor sea mayor a 3.5
df_filtrado = df[(df["sepal.width"] >= 3.5) & (df["sepal.length"] > 5)]
df_filtrado
```

```python
registros = len(df_filtrado)
registros
```

```python
agregados = df_filtrado.groupby("variety")["petal.width"].agg(["mean", "median", "std"])
agregados
```

```python
df_final = df_filtrado.loc[:, "petal.width":"variety"]
df_final["petal.width.st"] = (df_filtrado["petal.width"] - df_filtrado["petal.width"].mean()) / df_filtrado["petal.width"].std()
df_final = df_final.loc[:, "variety":"petal.width.st"]
df_final.rename(columns = {'variety':'Variedad', 'petal.width.st':'Ancho Est. Pétalo'}, inplace = True)
df_final

#dfp["IMM"] = dfp['Peso_KG'] / (dfp['Altura_CM'] / 100) ** 2
```
