```python
texto = input("Escribe una frase") 
conteo = {}
cadena = texto.split()
for nombre in cadena:
    if nombre not in conteo:
        conteo[nombre] = 1
    else:
        conteo[nombre] = conteo[nombre] + 1
for palabra, cantidad in conteo.items():
    print('La palabra', palabra,'se repite', cantidad, 'instancia(s)')
    
```
