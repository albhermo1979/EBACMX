```python
numero = input("Ingrese un número: ")

try:
    # Intentar convertir el número a entero
    n = int(numero)
    
    # Verificar si el número es negativo
    if n < 0:
        print("Error de captura")
    else:
        # Calcular el factorial
        factorial = 1
        for i in range(1, n + 1):
            factorial *= i
        print(f"Factorial de {n}: {factorial}")

except ValueError:
    print("Error de captura")

# Pruebas con números específicos
numeros = [4, 5, 6, 7.5]

for numero in numeros:
    try:
        n = int(numero)
        if n < 0:
            factorial = "Error de captura"
        else:
            factorial = 1
            for i in range(1, n + 1):
                factorial *= i
    except ValueError:
        factorial = "Error de captura"
    
    print(f"Factorial de {numero}: {factorial}")
```
