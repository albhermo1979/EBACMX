```python
import numpy as np
accidentes = np.array([[0,1,1],
                     [1,0,1],
                     [1,3,2],
                     [2,2,0]])
```

```python
#Suma de los accidentes (Suma del contenido del arreglo)
accidentes.sum()
```

```python
#La suma de las columnas
accidentes_areas = accidentes.sum(axis = 0)
accidentes_areas
```

```python
print("Los accidentes del area 1:", accidentes_areas[0] )
print("Los accidentes del area 1:", accidentes_areas[1] )
print("Los accidentes del area 1:", accidentes_areas[2] )
```

```python
#La suma de los renglones
accidentes_anios = accidentes.sum(axis = 1)
accidentes_anios
```

```python
#La suma de los renglones
accidentes_anios = accidentes.sum(axis = 1, keepdims=True) #Keepdims mantiene la dimensionde del arreglo
accidentes_anios
```

```python
print("Los accidentes del area 1:", accidentes_anios[0] )
print("Los accidentes del area 2:", accidentes_anios[1] )
print("Los accidentes del area 3:", accidentes_anios[2] )
print("Los accidentes del area 4:", accidentes_anios[3] )
```

```python
#Sacar los maximos y minimos
maximo = accidentes.max()
minimo = accidentes.min()
print(f"El numero maximo de accidentes es :{maximo}")
print(f"El numero minimo de accidentes es: {minimo}")
```

```python
#promedio 
promedio = accidentes.mean()
promedio
```

```python
#Operaciones vectorizadas
#Hay que hacer la suma del [1,2,3] + 3
#                          [4,5,6]
#Se define el arreglo
arreglo = np.array([[1,2,3],
                   [4,5,6]])
dimensiones= arreglo.shape
print(dimensiones[0]) #Renglones
print(dimensiones[1]) #Columnas
print(arreglo[0][0]) # Conocer el primer valor del arreglo
```

```python
#Hacer el ejercicio con un ciclo
import time
inicio = time.time()
arreglo_final = np.zeros((2,3))
for renglon in range(dimensiones[0]):
    for columnas in range(dimensiones[1]):
        arreglo_final[renglon][columnas] = arreglo[renglon][columnas] + 3
print(arreglo_final)
fin = time.time()
tiempo_final = inicio - fin
print(tiempo_final)
```
