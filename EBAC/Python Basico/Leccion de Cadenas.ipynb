```python
nombre1 = "Jose "
nombre2 = "Alberto"
nombre_completo = nombre1 + nombre2
print(nombre_completo)
```

```python
#El limite inferior de una cadena es 0
print(nombre1[0])
```

```python
#El limite superior de una cadena es el ultimo caracter mas 1
print(nombre1[0:3])
print(nombre1[0:5])
#el lo mismo asi
print(nombre1[:4])
print(nombre1[0:])
```

```python
#Para saber el numero de caracteres en una cadena
print(len(nombre_completo))
```

```python
#Para saber todo el tipo de comandos que pueden ser aplicados a una variable de cadena
dir(nombre1)
```
