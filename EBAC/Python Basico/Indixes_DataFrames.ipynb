```python
import numpy as np
import pandas as pd
#dfp = pd.DataFrame([['Bella','Labrador','Cafe', 56, 24, '2013-07-01'], ['Charlie','Poodle','Negro', 43, 24, '2016-09-16'], ['Lucy','Chow Chow','Café', 46, 24, '2014-08-35'], ['Cooper','Schnauzer','Gris', 49, 17, '2011-12-11'], ['Bernie','San Bernardo','Blanco', 77, 74, '2018-02-27'], ['Max','Chow Chow','Blanco', 45, 28, '2019-03-27'], ['Reed','Schnauzer','Negro', 45, 15, '2017-09-18']], index = [0, 1, 2, 3, 4, 5, 6], columns = ["Nombre", "Raza", "Color", "Altura_CM", "Peso_KG", "Fecha_Nac"])
arreglo = np.array([['Bella','Labrador','Cafe', 56, 24, '2013-07-01'], 
                    ['Charlie','Poodle','Negro', 43, 24, '2016-09-16'], 
                    ['Lucy','Chow Chow','Café', 46, 24, '2014-08-35'], 
                    ['Cooper','Schnauzer','Gris', 49, 17, '2011-12-11'], 
                    ['Bernie','San Bernardo','Blanco', 77, 74, '2018-02-27'], 
                    ['Max','Chow Chow','Blanco', 45, 28, '2019-03-27'], 
                    ['Reed','Schnauzer','Negro', 45, 15, '2017-09-18']])
arreglo
```

```python
#Para hacer Slicing del DataFrame con indices multiples usando Tuplas
dfp_ind2.loc[("Chow Chow", "Café"):("Schnauzer","Gris")]
```

```python
#Para hacer slicing de las columnas en DataFrame
dfp_ind3 = dfp_ind2.loc[:,"Nombre":"Peso_KG"]
dfp_ind3
```

```python
dfp_ind4 = dfp_ind2.loc[("Chow Chow", "Café"):("Schnauzer","Gris"),"Nombre":"Peso_KG"]
dfp_ind4
```

```python
#Creacion de la table pivote
dfp_altura_raza_color = dfp.pivot_table("Altura_CM", index = "Raza", columns = "Color")
dfp_altura_raza_color 
```

```python
#Extraccion de registros
dfp_altura_raza_color.loc["Labrador":"San Bernardo"]
```

```python
#Calculo del promedio para TODOS los renglones (por columna)
dfp_altura_raza_color.mean(axis = "index")
```
