```python
cadena_horas_laboradas = input("Introduce el Numero de horas trabajadas:")
cadena_tarifa_por_hora = input("Ingrese su tarifa de pago por hora: ")
try:
    numeroHoras = float(cadena_horas_laboradas)
    tarifa = float(cadena_tarifa_por_hora)
    if numeroHoras < 0 or tarifa < 0:
        print("parámetros incorrectos")
    else: 
        if numeroHoras <= 40:
            pago_total= numeroHoras * tarifa
            print("El monto de la tarifa a pagar es:", pago_total)
        if numeroHoras > 40 and numeroHoras <= 50:
            diferenciaHoras = numeroHoras - 40      
            tarifabasica = 40 * tarifa
            tarifaDiferencia = diferenciaHoras * tarifa * 1.1
            pago_total = tarifabasica + tarifaDiferencia 
            print("El monto de la tarifa a pagar es:", pago_total)
        elif numeroHoras > 50:
            exdecente40 = 10
            tarifabasica = 40 * tarifa
            diferenciaHoras = numeroHoras - 50
            tarifaDiferencia = diferenciaHoras * (tarifa * 1.2)
            pago_total = tarifabasica + (exdecente40 * tarifa * 1.1) + tarifaDiferencia 
            print("El monto de la tarifa a pagar es:", pago_total)
    
except:
 print("parámetros incorrectos")
```
