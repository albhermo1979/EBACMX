```python
k = int(input(print("Proporcina el numero a calcular: ")))
potencia = 0
i = 1
es_potencia_de_2 = False
while i <= k :
    potencia = 2 ** i
    if potencia == k:
        es_potencia_de_2 = True 
#        print(f"El numero {k} es potencia de 2")
        break
    else:
        i += 1
        potencia = 0
es_potencia_de_2
```
