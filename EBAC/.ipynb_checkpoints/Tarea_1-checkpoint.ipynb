```python
texto = input("Escribe una frase") 
conteo = {}
cadena = texto.split()
for nombre in cadena:
    if nombre not in conteo:
        conteo[nombre] = 1
    else:
        conteo[nombre] = conteo[nombre] + 1
for X in conteo:
    print('La palabra ', conteo.keys(X), 'Aparece', conteo.values(X), 'veces')
    
```
