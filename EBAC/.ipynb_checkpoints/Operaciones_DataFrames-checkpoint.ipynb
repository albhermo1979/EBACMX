```python
import pandas as pd
dfp = pd.DataFrame([['Bella','Labrador','Cafe', 56, 24, '2013-07-01'], ['Charlie','Poodle','Negro', 43, 24, '2016-09-16'], ['Lucy','Chow Chow','Café', 46, 24, '2014-08-35'], ['Cooper','Schnauzer','Gris', 49, 17, '2011-12-11'], ['Bernie','San Bernardo','Blanco', 77, 74, '2018-02-27'], ['Max','Chow Chow','Blanco', 45, 28, '2019-03-27'], ['Reed','Schnauzer','Negro', 45, 15, '2017-09-18']], index = [0, 1, 2, 3, 4, 5, 6], columns = ["Nombre", "Raza", "Color", "Altura_CM", "Peso_KG", "Fecha_Nac"])
dfp
```

```python
# Calcular la media de los pesos con el varlor redondeado a 2
dfp['Peso_KG'].mean().round(2)
```

```python
#Cual es la fecha mas antigua
dfp['Fecha_Nac'].min()
```

```python
#función de agregacion estadistica
def percentil80(columna):
    return columna.quantile(0.80)

dfp['Altura_CM'].agg(percentil80)
```

```python
#Agrupacion estadistica de varias columnas
dfp[['Altura_CM', 'Peso_KG']].agg(percentil80)
```

```python
#función de agregacion estadistica con percentil 90
def percentil90(columna):
    return columna.quantile(0.90)

# Se pueden hacer 2 funciones de agregacion en una sola linea 
dfp['Altura_CM'].agg(percentil90)
```

```python
dfp['Altura_CM']
```

```python
#Funcion de suma acumulativa
dfp['Altura_CM'].cumsum()
```

```python
#Funcion para toma el minimo de los valores 
dfp['Altura_CM'].cummin()
```

```python
#Conteo por columna en este caso se va a sacar el conteo por raza
dfp['Raza'].value_counts()
```

```python
#A la funcion count se le puede agregar la frecuencia relativa de los valores (% de aparicion por valor)
dfp['Raza'].value_counts(normalize=True)
```

```python
#Resumen por grupo: Mejor Opcion
dfp.groupby('Color')['Peso_KG'].mean()
```

```python
#Agrupacion por color en base a diversas estadisticas
dfp.groupby('Color')['Peso_KG'].agg([min, max, sum])
```

```python
#Agrupacion por color y raza en base a la media del peso
dfp.groupby(['Color', 'Raza'])['Peso_KG'].mean()
```
