```python
#Uso de Break
while True:
    linea = input("Escribe un Texto: ")
    if linea == "Fin":
        break
    print(linea)
print("Proceso Terminado")
    
```

```python
#Uso de Continue
while True:
    linea = input("Escribe un Texto: ")
    if linea[0] == "#":
        continue
    if linea == "Fin":
        break
    print(linea)
print("Proceso terminado")
    
```

```python
# Sumar los numeros hasta el 
n = int(input("Captura el valor de n: "))
suma = 0
i = 1
while i <= n:
    suma += i
    i += 1
print(f"La suma es igual a {suma}")
```

```python
numeros = (1,2,3,4,5,6,7,8,9)
valores = len(numeros)
impares = 0
pares = 0
contador = 1
while contador <+ valores:
    if  numeros[contador-1] % 2 == 0:
        pares += 1
    else:
        impares += 1
    contador +=1
print(f"El numero de valores pares es {pares}")
print(f"El numero de valores impares es {impares}")











    
```

```python
#Impresion de la serie de fibonacci
x, y = 0, 1 #Asignacion a las variables x y y los valores como aparecen en la lista
print(x)

while y < 50:
    print(y)
    x, y = y, x + y #Asignacion de los valores y de la suma de los valores
    
```
