```python
#Creacion de un histograma
iris["sepal.length"].hist()
```

```python
subintervalos = int(round(len(iris) ** 0.5, 0))
subintervalos
```

```python
#pasar los subintervalos al histograma
iris["sepal.length"].hist(bins = subintervalos)
```

```python
#Vamos a agrupar el DataFrame para poder graficar por tipo de flor
#Aca de calculo el promedio de la columna petal.lenght del DataFrame Iris que esta siendo agrupado por variety
iris_tipo = iris.groupby("variety")["petal.length"].mean()
iris_tipo
```

```python
#Grafico de barras
iris_tipo.plot(kind="bar")
```

```python
import pandas as pd
data = {'Producto': ['A','B','C','D', 'A'],
       'Ventas':[100, 150, 200, 130, 200],
        'Descuento': [5, 10, 0, 7, 0]}
df = pd.DataFrame(data)
df
```
