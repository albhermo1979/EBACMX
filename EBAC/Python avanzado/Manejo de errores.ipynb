```python
#Excepciones
x = 12
if x > 10:
    raise Exception('x no puede sobrepasar el valor de 10. El valor de x fue: {}'.format(x))
print(x)
```

```python
# Evaluacion de condiciones con Assert
# Si es que se cumple la condicion se genera el error
x = "Hola"
#assert x == "Hola"
assert x == "Chao"
```

```python
# Tambien se pueden evalyar las condiciones (assert) que causan el problema
# y luego si es qiue no se cumplen, entondces se les deja fuera

# importa la libreria "sys" que validara que tipo de plataforma tenemos
import sys
print ('Plataform = ', sys.platform)
assert ('windows' in sys.platform)
```

```python
#Manejo de excepciones con Try y Except
def linux_interaction():
    assert('Linux' in sys.platform), "Solo Linux"
    print('Aqui Casual')

try:
    linux_interaction()
except:
    print('Ambiente incorrecto')
```

```python
# Se modifica ahora el codigo del except con un "Assert"
# Se cacha la funcion Assertion en una variable
try:
    linux_interaction()
except AssertionError as error:
    print(error)
    print('La funcion de linux?interaction() no se ejecuto')
```

```python
# Otro ejemplo muy comun, la apertura de archivos
# Se trata de abrir un archivo (que sabemos que NO existe)
try:
    with open('file.log') as file:
        read_data = file.read()   
except:
    print('No puedo abrir el file.log')
```

```python
#ejemplo con else
try:
    num = int(input("ingrese un numero: "))
    assert num % 2 == 0
except:
    print('No ingresaste un numero par!')
else:
    reciprocal = 1/num
    print(reciprocal)
```

```python
#Ejemplo de division en cero
def divide(x, y):
    try:
        result = x // y
        print("Respuesta: ", result)
    except ZeroDivisionError:
        print('Division cero')

divide(7, 2)
divide(7, 0)
```

```python
#Ejemplo de division en cero con modificacion
def divide(x, y):
    try:
        result = x // y
    except ZeroDivisionError:
        print('Division cero')
    else:
        print("Respuesta: ", result)

divide(7, 2)
divide(7, 0)
```

```python
#Ejemplo de division en cero con modificacion
def divide(x, y):
    try:
        result = x // y
    except ZeroDivisionError:
        print('Division cero')
    else:
        print("Respuesta: ", result)
    finally:
        #Este se ejecuta siempre para cerrar un archivo o hacer liberacion de memoria
        print('Este siempre se ejecuta')

divide(7, 2)
divide(7, 0)
```

```python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: No se puede dividir por cero.")
    else:
        print("El resultado es:", result)
    finally:
        print("Operación completada.")

divide(10, 2)
divide(10, 0)
```
