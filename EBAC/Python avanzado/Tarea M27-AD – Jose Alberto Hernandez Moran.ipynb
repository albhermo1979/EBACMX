```python
#Ejemplo donde se busca un archivo ‚Äútest.txt‚Äù que NO existe en el directorio donde se est√° buscando.  
#Excepciones y las funciones TRY/EXCEPT 
try:
    with open('Test.txt') as file:
        read_data = file.read()   
except:
    print('No puedo abrir el archivo')
```

```python
#Ejemplos con TRY/EXCEPT/ELSE/FINALLY 
try:
    archivo = open("test.txt", "r")
except:
    print("‚ùå El archivo 'test.txt' no existe.")
else:
    print("‚úÖ Archivo abierto correctamente.")
    print(archivo.read())
finally:
    try:
        archivo.close()
        print("üîí Archivo cerrado.")
    except NameError:
        print("‚ÑπÔ∏è No se abri√≥ ning√∫n archivo, por lo tanto no se cerr√≥.")
```

```python
# Paso 1: Operaci√≥n matem√°tica
a, b, c, d = 5, 10, 15, 20
suma = a + b + c + d  # suma = 50

# Paso 2: Divisi√≥n por cero con manejo de excepci√≥n
try:
    resultado = suma / 0
except ZeroDivisionError as e:
    print(f"‚ö†Ô∏è Error de divisi√≥n por cero: {e}")
else:
    print(f"‚úÖ Resultado: {resultado}")
finally:
    print("üîö Fin del bloque de debugging.")
```
