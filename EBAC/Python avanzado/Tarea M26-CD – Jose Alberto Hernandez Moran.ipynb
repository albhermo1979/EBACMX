```python
#Abrir el archivo de lectura
file = open(r'./Height of Male and Female by Country 2022.csv')
csvreader = csv.reader(file)
```

```python
# Imprimir las primeras 50 lineas
# Voy a extraer los datos a una lista
rows = []
for row in csvreader:
    rows.append(row)

print(rows[:51])
```

```python
# Calcular la media, mínima y máxima altura de hombres y mujeres del data set completo 
from statistics import mean, median, mode, stdev
import numpy as np
alturas = np.array(rows)
alturas[0]
```

```python
#Primero saco en una lista aparte la Altura de los hombres y lo convierto a numeros flotantes
Male_Height_cm = alturas[1:,2]
Female_Height_cm = alturas[1:,2]
#Male_Height_cm
Male_Height_cm_num = [float(numeric_string) for numeric_string in Male_Height_cm]
Female_Height_cm_num = [float(numeric_string) for numeric_string in Female_Height_cm]
Male_Height_cm_num[0:10]
Female_Height_cm_num[0:10]
```

```python
# Calculo de la media minima y maxima de los hombres
mean_male = mean(Male_Height_cm_num)
min_male = min(Male_Height_cm_num)
max_male = max(Male_Height_cm_num)
mean_female = mean(Female_Height_cm_num)
min_female = min(Female_Height_cm_num)
max_female = max(Female_Height_cm_num)

#Imprimir los hallazgos
print('Los datos muestran que en promedio la altura de los hombres en centímetros es {:.2f}, y de las mujeres es {:.2f}'.format(mean_male, mean_female))
print('La altura minima en los hombres es {:.2f}, y de las mujeres es {:.2f}'.format(min_male, min_female))
print('Y la altura maxima en los hombres es {:.2f} y de las mujeres es {:.2f}'.format(max_male, max_female))
```

```python
# Filtar por mi pais
filtro = (alturas[:, 1] == 'Mexico')
arre_filtrado = alturas[filtro]
#print(arre_filtrado)
Male_Height_cm_num_mx = [float(numeric_string) for numeric_string in arre_filtrado[:, 2]]
Female_Height_cm_num_mx= [float(numeric_string) for numeric_string in arre_filtrado[:, 3]]
print('Los datos muestran que la altura de los hombres en centímetros es {:.2f}, y de las mujeres es {:.2f}'.format(Male_Height_cm_num_mx[0], Female_Height_cm_num_mx[0]))
```
