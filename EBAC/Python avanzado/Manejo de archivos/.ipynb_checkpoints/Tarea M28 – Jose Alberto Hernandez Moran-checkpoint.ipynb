# Leer Archivo CSV

```python
# Leer el archivo CSV y guardarlo en una lista
import csv
import os
import numpy as np

file = open(r'./fifa_eda.csv', encoding= 'UTF8')

csvreader = csv.reader(file)

header = []
header = next(csvreader)  #Leer la primera lista del archivo
rows = []
for row in csvreader:
    rows.append(row)
print(header)
print(rows[:3])
```

```python
# Para saber el largo del archivo
len(rows)
```

```python
#Pasar el archivo a una lista
fifas = np.array(rows)
```

```python
#Leemos la 10 primeras lineas
fifas[0:10]
```

```python
#Leemos las 10 ultimas lineas
fifas[-10:]
```

```python
#Gurdar los datos a un nuevo csv
header_nuevo_fifas = header
datos_nuevo_fifas = fifas[20:25]
#print(header_nuevo_fifas)
#print(datos_nuevo_fifas)
with open('nuevo_fifa.csv', 'w', encoding= 'UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header_nuevo_fifas)
    writer.writerow(datos_nuevo_fifas)
```

# Leer Archivo de excel

```python
#Leer el archivo de excel
import openpyxl
from pathlib import Path

xlsx_file = Path('.', 'Amazon_top100_bestselling_books_2009to2021.xlsx')
wb_obj = openpyxl.load_workbook(xlsx_file)
#Para abrir el sheet activo
sheet = wb_obj.active

print(sheet["C3"].value)
```

```python
#Para leer los rangos de las 10 primeras lineas
rango_celdas_10 = sheet['A1':'J10']
print(rango_celdas_10)
```

```python
#Para sacar los ultimos 10 renglones
total_filas = sheet.max_row
print(total_filas)
```

```python
#Imprimir los ultimos 10 renglones
for row in sheet.iter_rows(min_row=total_filas - 9, max_row=total_filas, values_only=True):
    print(row)
```

```python
#Guardar en un archivo de excel
wb_nuevo = openpyxl.Workbook()
sheet_nuevo = wb_nuevo.active
for i, fila in enumerate(rango_celdas_10, start=1):
    for j, celda in enumerate(fila, start=1):
        sheet_nuevo.cell(row=i, column=j).value = celda.value
wb_nuevo.save("Amazon_top10_rango.xlsx")
```

# Leer un archivo JSON

```python
#Lectura de un archivo tipo JSON
import json

#Se selecciona el archivo a abrir
with open('season1.json', 'r') as json_file:
    subtitulos_json = json.load(json_file) #Se carga el archivo
print(subtitulos_json)
```

```python
#Esto es para ver la estructura del archivo JSON
print(json.dumps(subtitulos_json, sort_keys=True, indent=4))
```

```python
#Para saber el nombre de las claves
print("üîç Claves del JSON:")
print(subtitulos_json.keys())
```

```python
primer_ep = list(subtitulos_json.keys())[0]
episodio = subtitulos_json[primer_ep]

# Ordenar las claves num√©ricamente (porque est√°n como strings)
claves_ordenadas = sorted(episodio.keys(), key=lambda x: int(x))

# Mostrar los primeros 10 subt√≠tulos
print(f"Primeros 10 subt√≠tulos de: {primer_ep}")
for i, clave in enumerate(claves_ordenadas[:10], start=1):
    subtitulo = episodio[clave]
    print(f"{i}. {subtitulo}")
```

```python
ultimo_ep = list(subtitulos_json.keys())[9]
episodio = subtitulos_json[ultimo_ep]

# Ordenar las claves num√©ricamente (porque est√°n como strings)
claves_ordenadas = sorted(episodio.keys(), key=lambda x: int(x))

# Mostrar los primeros 10 subt√≠tulos
print(f"Primeros 10 subt√≠tulos de: {ultimo_ep}")
for i, clave in enumerate(claves_ordenadas[:10], start=1):
    subtitulo = episodio[clave]
    print(f"{i}. {subtitulo}")
```

```python
#Construir un nuevo archivo con los 10 primeros subtitulos
primeros_10_subtitulos = {clave: episodio[clave] for clave in claves_ordenadas[:10]}
with open('primeros_10_subtitulos.json', 'w', encoding='utf-8') as salida:
    json.dump(primeros_10_subtitulos, salida, ensure_ascii=False, indent=4)

print("Archivo 'primeros_10_subtitulos.json' guardado correctamente.")
```
