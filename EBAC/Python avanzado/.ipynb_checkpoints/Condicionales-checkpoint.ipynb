```python
x = 5
if x == 5:
    print("El numero es igual a 5")
if x > 4:
    print("Mayor que 4")
if x >= 5:
    print("Mayor o igual que 5")
if x < 6:
    print("Menor que 6")
if x != 6: print("Diferente de 6")
```

```python
x = input("Dame un numero:")
if float(x) > 1:
    print("Mayor a 1")
    if float(x) < 100:
        print("Menor que 100")
print("Fin")
```

## El uso de try y except

cadena = input("Dame un numero positivo: ")
try:
    entero = int(cadena)
except:
    entero = -1
if entero > 0:
    print("El valor es correcto")
else:
    print("El valor es correcto")
