```python
#Uso de librería BeautifulSoup y requests.  Parseo de información (e.g. mercadolibre, despegar.com, etc.) 
import requests
from bs4 import BeautifulSoup
import pprint

# Para simular que somos humanos
headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Allow-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

url = 'https://www.liverpool.com.mx'

# Hacemos la petición
data = requests.get(url, headers=headers)

# Variable para guardar los articulos del sitio web
url_datos = []

html = BeautifulSoup(data.text, 'html.parser')

  
# buscamo stodos los enlaces de navegación o productos (etiqueta 'a')
articles = html.find_all('a', href=True)

# Recorremos los enlaces encontrados
for article in articles:
    
    # 1. Obtenemos el texto del enlace 
    title = article.get_text(strip=True)
    
    # 2. Obtenemos la URL a donde lleva el enlace 
    link_url = article['href']
    

    css_class = article.get('class', ['Sin clase'])[0]

    # Se si el título no está vacío
    if title:
        url_datos.append({
            "title": title, 
            "link": link_url, 
            "info_extra": css_class
        })

pprint.pprint(url_datos)
```
