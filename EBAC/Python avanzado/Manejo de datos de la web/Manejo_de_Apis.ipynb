# Obtener informacion desde una API
 * Las librerias se pueden llamar directamente con la libreria request
 * Codigos de retorno:
   - 200: OK, conexion saludable
   - 204: Se hizo conexion sin ningun datos
   - 401: Fallo la autenticacion
   - 403: Acceso prohibido 
   - 404: No se encuenra el server de la API
   - 500: Internal Server 

```python
#Importante empezar a programar cuando tengamos de respuesta un codigo 200
import requests
import pprint
response_API = requests.get('https://ebac.mx/')
response_API.status_code
```

```python
#Ejemplo
response = requests.get("http://api.open-notify.org/astros.json")
print(response.status_code)
```

```python
response = requests.get("http://api.open-notify.org/iss-now.json")
print(response.status_code)
pprint.pprint(response.json())
```

```python
#Para ir imprimiendo partes del archivo json
#print(response.json()['iss_position'])
print(response.json()['iss_position']['latitude'])
```

```python
lati_str = response.json()['iss_position']['latitude']
long_str = response.json()['iss_position']['longitude']
```

```python
type(lati_str)
```

```python
lati = float(lati_str)
long = float(long_str)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=long, y=lati)
plt.show
```

```python
#Enviar parameros  a las funciones
#En esta funcion en particular se pasa la latitud y la longitud, y la API indica cuantas vceses la estacion espacial ha pasado por ahi
#Aqui el ejemplo, esuando la longitud y la latitud de Ciudad de Mexico: 19.4326 N, 99.1332 W
import json
#se genera una funcion simple llamada "jprint"
def jprint(obj):
    #Create a formatted string of the python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)


parameters = {
    "lat": 19.43,
    "lon": -99.13
}
response = requests.get("http://api.open-notify.org/iss-now.json", params=parameters)
print(response.status_code)
jprint(response.json())
```
