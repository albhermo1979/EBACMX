# No existen columnas con valores numericos que sean objects

```python
#Añadir una columna “Years Playing”, que calcule el año actual menos la columna “Joined”
from datetime import datetime
current_y = datetime.now().year

df_fifa['Years Playing'] = current_y - df_fifa['Joined']
df_fifa.head(10)
```

```python
#Buscar y mostrar a todos los jugadores de Mexico
df_fifa[['Name','Nationality','Club']].loc[df_fifa['Nationality'] == 'Mexico']

```

```python
#Ordenar y mostrar los datos por la columna ReleaseClause (sueldo)
df_fifa.sort_values(by = "Release Clause", ascending=False)
```

```python
#Generar un nuevo dataset que contenga el año (joined) y el número de jugadores (groupby)
df_fifa_count_by_year = df_fifa.groupby("Joined").count()['Name']
df_fifa_count_by_year
```

```python
#❖Opcional❖: Generar un gráfico que contenga, por año, el número de jugadores
df_fifa_count_by_year.plot.bar()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Si df_fifa_count_by_year es una Serie
df_plot = df_fifa_count_by_year.reset_index()  
df_plot.columns = ["Year", "Count"]   # renombrar columnas

sns.barplot(data=df_plot, x="Year", y="Count", palette="mako")
plt.show()

```
