## Insertando, Eliminando, Edicion de Dataframe nuevo

```python
#Nuevo Dataframe
clase = [('Jorge', 22, 'MX'),
        ('Anita', 31, 'MX'),
        ('Raul', 16, 'BR'),
        ('Simon', 41, 'BR'),
        ('Diego', 33, 'PA'),
        ('Jorge', 35, 'PA'),
        ('Almendra', 35, 'COL'),
        ('Mary', 35, 'COL'),
        ('Peter', 33, 'EC'),]

df_clase = pd.DataFrame(clase, columns=['Nombre', 'Edad', 'Pais'])
df_clase
```

```python
#ordernar la clase por edad
df_clase.sort_values(by = 'Edad')
```

```python
#ordenar la clase por edad
df_clase.sort_values(by = 'Edad', ascending=False)
```

```python
#Lega un alumno nuevo (se pone la opcion index para que se añada el indice automaticamente)
#df_clase = df_clase.append ({'Nombre':'Isabel', 'Edad':35, 'Pais':'PER'}, ignore_index=True)
#La funcion append ya no funciona con Pandas 2.0 hay que usar Concat
#Hay que generar un dataframe nuevo con el nuevo registro y luego concatenarlo
nuevo = [{'Nombre':'Isabel', 'Edad':35, 'Pais':'PER'}]
df_nuevo = pd.DataFrame(nuevo, columns=['Nombre', 'Edad', 'Pais'])
df_clase = pd.concat([df_clase, df_nuevo], ignore_index=True)
df_clase
```

```python
#tambien se puede usar
#df.loc[len(df)] = ['Paulo', 48, 'COL']
```

```python
#Eliminacion de registro
#Si solo lo hacemos asi la funcion se ejecuta pero no hace nada, porque se tiene que asignar la funcion al Dataframe
df_clase.drop(df_clase.index[9])
```

```python
#Esta es la forma correcta de hacerlo
df_clase = df_clase.drop(df_clase.index[9])
```

```python
df_clase
```

```python
#Eliminar mas de un registro a la vez
#tener cuidado siempre con el doble corchete
df_clase.drop(df_clase.index[[7,8]])
```

```python
df_clase = df_clase.drop(df_clase.index[[7,8]])
df_clase
```

```python
#Para modificar un registro
#Se puede localiozar con .at
df_clase.at[3, 'Nombre'] = 'Jose'
df_clase
```

```python
#Para mas de un registro se puede utilizar .loc y adenmas hacerlo para multiples columnas
df_clase.loc[1:2, ['Edad','Pais']] = [32, 'ARG']
df_clase
```

```python
#Se puede utilizar replace
df_clase.replace("PA", "PAN", inplace=True)
```
