```python
df_fifa.drop(df_fifa[df_fifa['Nationality'] == 'St Lucia'].index, inplace=True)
```

```python
#df['Nationality'].unique()
#Me servia para revisar como estaba el nombre en el DataFrame
df_fifa[df_fifa['Nationality'].str.startswith('St L', na=False)]
```

```python
#Generar una lambda que añada la zona horaria y la capital del país de nacionalidad
import pytz
import pycountry
df_fifa['Country_Code'] = df_fifa['Nationality'].apply(lambda x: pycountry.countries.lookup(x).alpha_2 if pycountry.countries.lookup(x) else None)
```

```python
df_fifa
```

```python
#Agrego la zona Horaria
df_fifa['ZonaHoraria'] = df_fifa['Country_Code'].apply(lambda x: pytz.country_timezones[x])
```

```python
#Agrego la capital del pais
from countryinfo import CountryInfo
df_fifa['Capital'] = df_fifa['Country_Code'].apply(lambda x: CountryInfo(x).capital()).head(1)
```

```python
df_fifa.head()
```

```python
#Generar una lambda que en base al potencial, añada una columna “Candidato” o “Next Year”, para el premio para el balón de oro.  Jugadores con más de 85, son candidatos.
df_fifa['BalonOro'] = df_fifa['Potential'].apply(lambda x: 'Candidato' if x > 85 else 'Next Year')
```
