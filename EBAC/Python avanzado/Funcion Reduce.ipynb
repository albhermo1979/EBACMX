```python
#Funcion Reduce
# La función reduce() en Python aplica una función acumulativa a los elementos de un iterable, reduciéndolos a un solo valor.
#Ejemplo de suma simple
def sum(a, b):
    result = a + b
    print(f"{a} + {b} = {result}")
    return result

sum(5, 10)
```

```python
#Para que funcione la funcion reduce se debe de importar la libreria functools
from functools import reduce

#Se genera la lista de numeros a ser procesados
numeros = [0, 1, 2, 3, 4]

#Se aplica la funcion de suma sobre el listado de numeros
reduce(sum, numeros)

```

```python
from functools import reduce

numeros = [0, 1, 2, 3, 4]

reduce(sum, numeros, 100)
```

```python
#Ejemplo de suma de valores numericos

numbers = [0, 1, 2, 3, 4]
total = 0
for num in numbers:
    total += num
total
```

```python
#Funcion para sumar numeros
def my_add (a, b):
    return a+b
my_add(1,2)
```

```python
from functools import reduce

numeros = [0, 1, 2, 3, 4]

reduce(my_add, numeros)
```

```python
#La misma funcion aterior pero con lambda
from functools import reduce

numeros = [0, 1, 2, 3, 4]
reduce(lambda a, b: a + b, numeros)
```

```python
#Usando reduce para calcular maximos y minimos
#Se aplican las funciones minimo y maximo a todos los items de la list
from functools import reduce

#Funcion para calcular minimo
def minimo (a, b):
    if (a<b):
        return a
    return b

#funcion para calcular maxico
def max(a, b):
    if (a>b):
        return a
    return b
    #return a if a > b else b #Esta es otra forma de hacer lo de la lineas anteriores

numeross = [4, 5, 9, -20, 15, 109, 25, -1]

mini = reduce(minimo, numeross)
maxi = reduce(max, numeross)

#Se imprime el resultado de la lista
print(f'Minimo = {mini}')
print(f'Maximo = {maxi}')
```
