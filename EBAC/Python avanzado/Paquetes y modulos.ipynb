```python
#Se importa el modulo de numpy para generar ese tipo de analisis
import numpy as np
```

```python
movies = np.array(rows)
movies[0]
```

```python
#Ahora le vamos a pedir que nos de el anio de las 50 primeras peliculas
movies[0:49,7] #Los primeros 2 numeros son el rango y el ultimo es el numero de la columa 
```

```python
#Convertir la columna del a;o en valores (ahora es un string)
type(movies[0,7])
```

```python
#Vamos a generar un arreglo para trabajar sobre el
#El : trae todas las lineas de la columna 7
release_year = movies[:,7]
release_year
```

```python
#Transformacion de un arreglo a numeros
release_year_int = [int(numeric_string) for numeric_string in release_year]
release_year_int[0:10]
```

```python
#Primer grafico, una distribucion de los anios
import matplotlib.pyplot as plt
plt.hist(release_year_int)
plt.show()
```

```python
#Subimos el numero de bins
plt.hist(release_year_int, bins=70)
plt.show()
```

```python
#Generamos un titulo
plt.hist(release_year_int, bins=70)
plt.title('Histograma de numero de peliculas por ao - BD Netflix')
plt.show()
```

```python
# Calculo de la media y desviacion estandar del arreglo
mean_year = mean(release_year_int)
stdev_year = stdev(release_year_int)
min_year = min(release_year_int)
max_year = max(release_year_int)
#release_year_int[20:30]

#Primer reporte estadistico de los datos
print('En promedio, el ao en el que se generarion las peliculas del archico es el', int(mean_year))
print('Ademas, la desviacion estandar de los datos nos muestra que se tienen', int(stdev_year), 'de desviacion estandard')
print('El primer ao que se muestra en los datos es', int(min_year))
print('El ultimo ao que se muestra en los datos es', int(max_year))
```

```python
# Como el ao no es un dato que nos de mucho para hacer un analisis de datos se va a tomar la columna de duracion.
# La columna de duracion se divide en min o en season.
# Para este analisis SOLO se van a tomar en cuenta los minutos
# Asignamo un array duration a las duraciones de todas las peliculas
duration = movies[0:,9]
duration [0:10]
```

```python
#Se obtienen las peliculas que vienen en minutos
movies[np.char.endswith(movies, 'min')][:9]
```

```python
#Generar un arreglo de las duraciones en minutos
duration_min = (duration[np.char.endswith(duration, 'min')])
duration_min
```

```python
# Una vez que se tiene el arreglo de solo los minutos se convierte el registro de: "90 min" a solo "90"
# Esto se hace utilizando la funcion "strip" de la libreria de numpy (np.char)
# Elimina la terminacion min
duration_min_end = np.char.strip(duration_min, 'min')
# Despues se transforma a numeros
duration_min_int = [int(numeric_string) for numeric_string in duration_min_end]
duration_min_int[0:9]
```

```python
# Calculo de la media y desviacion estandar del arreglo de duracion de las peliculas
mean_movie = mean(duration_min_int)
stdev_movie = stdev(duration_min_int)
min_movie = min(duration_min_int)
max_movie = max(duration_min_int)
#release_year_int[20:30]

#Primer reporte estadistico de los datos
print('En promedio, la duracion de las peliculas es de', int(mean_movie), 'min')
print('Ademas, la desviacion estandar de los datos nos muestra que se tienen', int(stdev_movie), 'de desviacion estandard')
print('La pelicula con menos minutos que se muestra en los datos es', int(min_movie))
print('La pelicula con mas duracion en min que se muestra en los datos es', int(max_movie))
```

```python
#Nuevos graficos
plt.title('Histograma de duracion de las peliculas')
plt.hist(duration_min_int, bins=70)
plt.show()
```
