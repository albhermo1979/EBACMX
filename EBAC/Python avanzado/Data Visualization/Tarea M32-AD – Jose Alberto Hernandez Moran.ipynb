```python
# Lectura del archivo 
import pandas as pd
df = pd.read_csv('fifa_eda.csv')
df.head(5)
```

```python
# Elemento 1: Obtener la matriz de correlacion y dejarla como heatmap
#df.info()
# Como me marcar error al momento de usar la funcion CORR, solo tomo en cuenta loa valores numericos
import plotly.express as px
df_numericos = df.select_dtypes(include=['int64', 'float64'])
matriz_correlacion = df_numericos.corr()
```

```python
#Muestro el resultado de la matiz
matriz_correlacion
```

```python
# Creo el Heatmap
fig = px.imshow(matriz_correlacion, color_continuous_scale="RdBu", title="Matriz de correlación - Variables numéricas (FIFA)", aspect="auto")
fig.show()
```

# Explicacion de la salida:
  * La correlacion que hay entre la edad y el overall es de 0.452 que digamos que es media baja (mas cercana al cero).
  * Pero la grafica muestra que a partir de los 24 a;os los jugadores tienden a subir el overall

```python
# Elemento 3: Generar un gráfico de barras por club que indique el número de jugadores 
df_jugadores_club = df.groupby(['Club']).size().rename('Numero').reset_index()
```

```python
df_jugadores_club.head()
```

```python
#En este caso inclui un sample 20 para que se pudiera ver mejor
fig = px.bar(df_jugadores_club.sample(20) , x = "Club", y="Numero", title="Numero de jugadores por club Sample 20", text="Numero", color="Club")
fig.show()
```

```python
# Elemento 4: Hacer un gráfico multipanel que indique la relación altura (height) vs skill moves, siendo la variable del panel si es zurdo o derecho
fig = px.strip(df, x="Height", y="Skill Moves", facet_col="Preferred Foot", title="Relacion Altura vs Skill moves")
fig.show()
```

```python
# Elemento 5: Un gráfico que crea importante para mostrar a los potenciales cracks
# Para esto seleecione el Desempe;o, el potencial y la edad
Seleccion = ["Overall", "Potential", "Age"]
cracks_potenciales = df[Seleccion].dropna()

fig = px.scatter(
    cracks_potenciales, x="Overall", y="Potential", color="Age", color_continuous_scale="Viridis", opacity=0.6, title="Mapa de cracks potenciales (Overall vs Potential)", labels={"Age": "Edad"})
fig.show()
```
