```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv('netflix_titles.csv')
```

```python
df.sample(5)
```

```python
import plotly.express as px
import numpy
```

```python
#Con esto se agrupa pero a la vez le doy un nombre a la columna resultante para que sea facil de llamar ('movies') y ademas se resetea el indice
df_movies_year = df.groupby('release_year').size().rename('movies').reset_index()
```

```python
df_movies_year
```

```python
#Grafico de barras
#Numero de peliculas y series por a;o
# Uso de Plotly.bar
fig = px.bar(df_movies_year, x = "release_year", y="movies", title="Series y peliculas por a;o")
fig.show()
```

```python
#Le vamos a agregar el type para que me distinga entre series y peliculas
df_movies_year_type = df.groupby(['release_year', 'type']).size().rename('movies').reset_index()
df_movies_year_type
```

```python
fig = px.bar(df_movies_year_type, x = "release_year", y="movies", title="Series y peliculas por a;o", color="type")
fig.show()
```

```python
#Se modifica el grafico para que se pongan barras agrupadas una al lado de otras
fig = px.bar(df_movies_year_type, x = "release_year", y="movies", title="Series y peliculas por a;o", color="type", barmode="group")
fig.show()
```

```python
# Hagamos un zoom desde 2000 a la fecha
# Se eliminan los registros desde el inicio hasta 1999
# Con la instruccion index lo que se saca es el numero de renglon que cumple con la condicion y luego lo uso para afectar el dataframe
index_1990 = df_movies_year_type[(df_movies_year_type['release_year'] < 2000)].index
df_movies_year_type.drop(index_1990, inplace=True)
```

```python
df_movies_year_type.head(3)
```

```python
fig = px.bar(df_movies_year_type, x = "release_year", y="movies", title="Series y peliculas por a;o desde el 2020", color="type", barmode="group", text="movies")
fig.show()
```

```python
#Grafico de lineas
#Numero de peliculas y series por a;o
# Uso de Plotly.bar
fig = px.line(df_movies_year_type, x = "release_year", y="movies", title="Series y peliculas por a;o", color="type")
fig.show()
```

```python
#Le vamos a agregar el raiting
df_release_raiting = df.groupby(['release_year', 'rating']).size().rename('movies').reset_index()
df_release_raiting.head(10)
```

```python
#Grafico de lineas por raiting
#Numero de peliculas y series por a;o
# Uso de Plotly.bar
fig = px.line(df_release_raiting, x = "release_year", y="movies", title="Series y peliculas por a;o", color="rating")
fig.show()
```

## Grafico de Pie
 * Que tipo de pelicula
 * Que tipo de pelicula (por raiting) se incluyeron, en total

```python
#Limpieza de ratings
df_release_raiting_total = df.groupby(['rating']).size().rename('movies').reset_index()
df_release_raiting_total
```

```python
df_release_raiting_total.drop([0, 1, 2], inplace=True)
```

```python
df_release_raiting_total = df_release_raiting_total.sort_values(by='movies')
```

```python
df_release_raiting_total 
```

```python
#Grafico de pie
fig = px.pie(df_release_raiting_total , names="rating", values="movies", title="Series y peliculas por a;o")
fig.show()
```
