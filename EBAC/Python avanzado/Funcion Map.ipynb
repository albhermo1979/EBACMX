#Funcion Map

```python
#Generando una operacion a ser aplicada a un listado de numeros
#Este en un ejemplo sin map

#Declaracion de los numeros a ser usados
org_list = [1, 2, 3, 4, 5]
fin_list = []

#Aplica la operacion a cada uno de los numeros de al listado
#** es para elevar al cubo
for num in org_list:
    fin_list.append(num**3)

#Imprime la lista
print(fin_list)
```

```python
#Ejemplo usando la funcion map()
#Define la lista inicial de numeros. No es necesario definir la lista de resultados
org_list = [1, 2, 3, 4, 5]

#Se define la funcion a ser aplicada a cada numero
def cube(num):
    return num**3

#Aplica la funcion map
fin_list = list(map(cube, org_list))
print(fin_list)
```

```python
#Programa que saca el largo de cada elemento en una vista
#Utilizando la funcion de Python "len"
org_list = ["Hola", "mundo", "EBAC"]
fin_list = list(map(len, org_list))
print(fin_list)
```

```python
#Funcion Map utilizando mas de un iterable (base y power)
base = [10, 10, 10, 10]
power = [1, 2, 3, 4]

result = list(map(pow, base, power))
print(result)
```

```python
#Ejemplo de map usando lambdas
org_list = [1, 2, 3, 4, 5]

fin_list = list(map(lambda x:x**3, org_list))
print(fin_list)
```

```python
#Llevar al doble un numero usando map y lambda
numbers = (1, 2, 3, 4)
result = map(lambda x: x + x, numbers)
print(list(result))
```

```python
#Suma de 3 listas
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
numbers3 = [7, 8, 9]

result = map(lambda x, y, z: x + y + z, numbers1, numbers2, numbers3)
print(list(result))

```

```python
#Ejemplo para descomponer cadenas

lista = ['sat', 'bat', 'cat', 'mat']

test = list(map(list, lista))
print(test)
```
