# Modelo de Presupuesto de Costos
 * Se generan algunas variables para el calculo de un presupuesto simple
 * Cada variable se sumara con la siguiente y sera una distribucion triangular 100, 150, 130
 * Esto quiere decir que para uno de los conceptos se tendra una variable que oscila de 100 a 300, para cada escenario

```python
# Variables de configuracion
num_simulaciones = 10000
max_num_rolls = 1000
pet = 1

#Variables del presupuesto
costo_estudios = 0
costo_diseno = 0
costo_ingenieria = 0
costo_implementacion = 0
costo_pruebas = 0
costo_aceptacion = 0
costo_referencia = 900
```

```python
# Se genera una funcio que "rueda los dados" - es decir saca una muestra de todas las distribuciones simultaneamente
# Se incluye la distribucion para cada variabe
def roll_dice():
    costo_estudios = np.random.triangular(100, 150, 300)
    costo_diseno = np.random.triangular(100, 150, 300)
    costo_ingenieria = np.random.triangular(100, 150, 300)
    costo_implementacion = np.random.triangular(100, 150, 300)
    costo_pruebas = np.random.triangular(100, 150, 300)
    costo_aceptacion = np.random.triangular(100, 150, 300)

    return costo_estudios, costo_diseno, costo_ingenieria, costo_implementacion, costo_pruebas, costo_aceptacion
```

```python
# Prueba 1 itereccion del presupuesto
a = roll_dice()

# Se ve como se retornan valores en un array
print(a[0], a[1], a[2], a[3], a[4], a[5])
```

```python
# Inputs
num_simulations = 1000
max_num_rolls = 1000
bet = 1

# Tracking
end_balance = []
```

```python
# Un ciclo para correr el numero de simulaciones deseadas
for i in range(num_simulations):
    balance = [900]
    num_rolls = [0]
    num_wins = 0
    # Corre las 100 veces
    while num_rolls[-1] < max_num_rolls:
        iter = roll_dice()
        #Obtiene el costo total
        costo_total = iter[0] + iter[1] + iter[2] + iter[3] + iter[4] + iter[5]
        balance.append(costo_total)
        num_rolls.append(num_rolls[-1] + 1)
#PLotea las muestras del balance
plt.plot(num_rolls, balance)
```

```python
# Imprime la media
np.mean(balance)
```

```python
# genera el histograma del balance
np.random.seed(42)
plt.hist(balance, density=True, bins=80)
plt.ylabel('Frecuencia')
plt.xlabel('Budget')
plt.axvline(np.mean(balance), color='k', linestyle='dashed', linewidth=1)
plt.axvline(costo_referencia, color='k', linestyle='dashed', linewidth=1)
```

```python
# Numero de muestras
print('Numero de muestras: ', balance.__len__())
total_muestras = balance.__len__()
```

```python
# Genera una lista con las muestras menores al costo de referencia en costo_ok
costo_ok = [x for x in balance if x < costo_referencia]
```

```python
# Probabilidad de que sea un total menor al costo de referencia
num_costo_ok = costo_ok.__len__()
prob_costo_ref = num_costo_ok / total_muestras
print(f'La probabilidad de que el costo sea menor al costo de referencia es: {prob_costo_ref*100:.2f} %')
```

# Cambios en las distribuciones de probabilidad
 * Se usan las distribuciones de probabilidad: normal, uniforme
 * Otras distribuciones a revisar: PERT, lognromal, etc
 * Esto es muy comun en modelo de este tipo, donde se encuemtra variables que responden a distintos tipos de distribucion

```python
# Se genera una funcion que rueda los dados
def roll_dice():
    costo_estudios = np.random.normal(100, 20)
    costo_diseno = np.random.triangular(100, 150, 300)
    costo_ingenieria = np.random.triangular(100, 150, 300)
    costo_implementacion = np.random.uniform(100, 300)
    costo_pruebas = np.random.uniform(200, 250)
    costo_aceptacion = np.random.triangular(100, 150, 300)

    return costo_estudios, costo_diseno, costo_ingenieria, costo_implementacion, costo_pruebas, costo_aceptacion
```

```python
# Prueba 1 itereccion del presupuesto
a = roll_dice()

# Se ve como se retornan valores en un array
print(a[0], a[1], a[2], a[3], a[4], a[5])
```

```python
# Inputs
num_simulations = 1000
max_num_rolls = 1000
bet = 1

# Tracking
end_balance = []
```

```python
# Un ciclo para correr el numero de simulaciones deseadas
for i in range(num_simulations):
    balance2 = [900]
    num_rolls = [0]
    num_wins = 0

    while num_rolls[-1] < max_num_rolls:
        iter = roll_dice()
        #Obtiene el costo total
        costo_total = iter[0] + iter[1] + iter[2] + iter[3] + iter[4] + iter[5]
        balance2.append(costo_total)
        num_rolls.append(num_rolls[-1] + 1)
#PLotea las muestras del balance
plt.plot(num_rolls, balance2)
```

```python
# Como comparar las correlaciones entre variables y graficarlas

# for loop to run for the number of simulations desired
for i in range(num_simulations):
    balance2 = [900]
    num_rolls=[0]
    num_wins=0

    #Array
    cestudios = []
    cdiseno = []
    cingenieria = []
    cimplementacion = []
    cpruebas = []
    caceptacion = []
    while num_rolls[-1] < 1000:
        iter = roll_dice()
        costo_total = iter[0] + iter[1] + iter[2] + iter[3] + iter[4] + iter[5]

        cestudios.append(iter[0])
        cdiseno.append(iter[1])
        cingenieria.append(iter[2])
        cimplementacion.append(iter[3])
        cpruebas.append(iter[4])
        caceptacion.append(iter[5])

        balance2.append(costo_total)
        num_rolls.append(num_rolls[-1] + 1)




```
