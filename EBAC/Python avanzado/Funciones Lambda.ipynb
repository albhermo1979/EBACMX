##Funciones Lambda

```python
# Ejempo a ser replicado
def add_one(x):
    return x + 1

add_one(101)
```

```python
#Con funciones lambda
add_one = lambda x: x +1
add_one(11)
```

```python
#ejemplo de suma simple
x = lambda a: a +  100
print(x(120))
```

```python
#Ejemplo de multiplicacion
x = lambda a, b: a * b
print(x(10, 60))
```

```python
# Se puede tener cualquier numero de argumentos
x = lambda a, b, c : a + b + c
print(x(10, 20, 30))
```

```python
# Se pueden usar funciones lambda dentro de otras funciones
def myfunc(n):
    return lambda a : a * n

#Generar la funcion para multiplicar el 2 por el numero
mydoubler =  myfunc(2)

print(mydoubler(10))
#en este caso el 10 es la a de lambda y el 2 es el que se le pasa a la funcion
```

```python
def myfunc(n):
    return lambda a : a * n

mytripler = myfunc(3)

print(mytripler(11))    
```

```python
#Se puede reutilizar la funcion
def myfunc(n):
    return lambda a : a * n

mydoubler =  myfunc(2)
mytripler = myfunc(3)

print(mydoubler(10))
print(mytripler(11))    
```

```python
#Ejemplo con strings
# title() se utiliza para que cada palabra empiece con mayusculas
full_name = lambda first, last: f'Full name: {first.title()} {last.title()}'
full_name('EBAC', 'curso de data Analisys')
```

```python
#Ejemplo de una lambda invocada de manera inmediata
(lambda x, y: x * y)(60, 10)
```

# Las funciones lambda no pueden tener statements dentro, como return, pass, assert, raise, etc
# Se escriben solo como una linea de ejecucion
# No soportan ningun tipo de anotaciones(comemtarios)
# Se pueden invocar inmediatamente
